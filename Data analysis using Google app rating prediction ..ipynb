Fill the null values with appropriate values using aggreagate functions such as mean , median or mode :

```python
# define a function impute_median
def impute_median(series):
    return series.fillna(series.median)
```

```python
app_data.isnull().sum()
```

```python
# Modes of Categorical values
print(app_data['Type'].mode())
print(app_data['Current Ver'].mode())
print(app_data['Android Ver'].mode())

```

```python
# Fill the missing categorical value modes
app_data['Type'].fillna(str(app_data['Type'].mode().values[0]), inplace=True)
app_data['Type'].fillna(str(app_data['Current Ver'].mode().values[0]), inplace=True)
app_data['Type'].fillna(str(app_data['Android Ver'].mode().values[0]), inplace=True)
```

```python
# To count no of null values in each column .
app_data.isnull().sum()
```

```python
app_data['Price'] = app_data['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
app_data['Reviews'] = pd.to_numeric (app_data ['Reviews'],errors='coerce')
```

```python
app_data['Installs'] = app_data['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
app_data['Installs'] = app_data['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))



```

# Data Visulaization 

```python
grp = app_data.groupby('Genres')
x = grp['Rating'].agg(np.mean)
y = grp['Reviews'].agg(np.mean)

print(x)
print(y)

```

```python
plt.figure(figsize=(20,8))
plt.plot(x,'ro')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(25,15))
plt.plot(x,'ro',color='g')
plt.xticks(rotation=90)
plt.title('Category wise rating')
plt.xlabel('Categories -->')
plt.ylabel('Ratings -->')
plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.plot(y,'r--',color='y')
plt.xticks(rotation=90)
plt.title('Category wise rating')
plt.xlabel('Categories -->')
plt.ylabel('Ratings -->')
plt.show()
```
